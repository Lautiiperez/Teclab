```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

poblacion = pd.read_csv('C:/Users/Lautaro Perez/Documents/Teclab/Analisis_y_Visualizacion_de_Datos/ARCHIVOS/poblacion.csv', encoding='latin-1')
esperanza = pd.read_csv('C:/Users/Lautaro Perez/Documents/Teclab/Analisis_y_Visualizacion_de_Datos/ARCHIVOS/esperanza_de_vida.csv', encoding='latin-1')
hogares = pd.read_csv('C:/Users/Lautaro Perez/Documents/Teclab/Analisis_y_Visualizacion_de_Datos/ARCHIVOS/hogares_viviendas_superficie.csv', encoding='latin-1')
empleo = pd.read_csv('C:/Users/Lautaro Perez/Documents/Teclab/Analisis_y_Visualizacion_de_Datos/ARCHIVOS/empleo.csv', encoding='latin-1')
exportaciones = pd.read_csv('C:/Users/Lautaro Perez/Documents/Teclab/Analisis_y_Visualizacion_de_Datos/ARCHIVOS/exportaciones.csv', encoding='latin-1')
```

```python
plt.figure(figsize=(10,6))
sns.histplot(poblacion['poblacion_total'], bins=30, kde=True)
plt.title('Distribución de la Población Total')
plt.xlabel('Población Total')
plt.ylabel('Frecuencia')
plt.show()
```

Muestra la desigual distribución de la población entre provincias.
La mayoría de las provincias tienen poblaciones bajas, con pocas provincias altamente pobladas.

```python
# Unir las tablas por provincia y año
df_empleo_poblacion = pd.merge(empleo, poblacion, on=['provincia', 'anio'])

# Gráfico de dispersión
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_empleo_poblacion, x='empleados_registrados_miles', y='poblacion_total')
plt.title('Relación entre Empleados Registrados y Población Total')
plt.xlabel('Empleados Registrados (en miles)')
plt.ylabel('Población Total')
plt.show()
```

Revela una correlación entre población y empleo registrado.
A mayor población, mayor cantidad de empleados registrados.

```python
plt.figure(figsize=(10,6))

# Evolución de esperanza de vida de mujeres
sns.lineplot(data=esperanza, x='anio', y='mujeres', label='Mujeres')

# Evolución de esperanza de vida de varones
sns.lineplot(data=esperanza, x='anio', y='varones', label='Varones')

plt.title('Evolución de la Esperanza de Vida')
plt.xlabel('Año')
plt.ylabel('Esperanza de Vida (años)')
plt.legend()
plt.show()
```

Refleja mejoras continuas en la salud y la diferencia entre géneros.
La esperanza de vida de mujeres y varones aumenta con el tiempo, siendo mayor en mujeres.
